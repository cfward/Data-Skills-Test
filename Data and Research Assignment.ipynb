```python
import pandas as pd

# Load the CSV data into a Pandas DataFrame
data = pd.read_csv("Skills Test Data Set - CanvasserData.csv")

# 1. How many unique canvassers did the organization have during their canvassing program?
unique_canvassers = data['Canvasser'].nunique()

# 2. How many shifts did each canvasser work?
shifts_worked = data['Canvasser'].value_counts().reset_index()
shifts_worked.index = shifts_worked.index + 1  # Start the index at 1
shifts_worked.columns = ['Canvasser', 'Shifts Worked']

# 3. How many doors did each canvasser knock and what were the results of those knocks?
doors_knocked_results = data.groupby('Canvasser').sum().reset_index()
doors_knocked_results.index = doors_knocked_results.index + 1  # Start the index at 1
doors_knocked_results.columns = ['Canvasser', 'Not Home', 'Refused', 'Moved', 'Deceased', 'Canvassed']

# 4. What is the contact rate for each canvasser?
# Assuming 'Not Home', 'Refused', 'Moved', and 'Deceased' are considered unsuccessful knocks
data['Unsuccessful Knocks'] = data['Not Home'] + data['Refused'] + data['Moved'] + data['Deceased']
contact_rate_df = data.groupby('Canvasser').apply(lambda x: (x['Canvassed'] / (x['Canvassed'] + x['Unsuccessful Knocks'])).mean()).reset_index()
contact_rate_df.index = contact_rate_df.index + 1  # Start the index at 1
contact_rate_df.columns = ['Canvasser', 'Contact Rate']

# 5. Recommendations to the field lead based on your analysis of this list:
# You can provide recommendations based on the analysis of the data and goals of the canvassing program.

# Print the results
print(f"1. Unique Canvassers: {unique_canvassers}")
print("\n2. Shifts Worked:")
print(shifts_worked)
print("\n3. Doors Knocked and Knock Results:")
print(doors_knocked_results)
print("\n4. Contact Rate:")
print(contact_rate_df)
```

```python
import pandas as pd

# Load the CSV data into a Pandas DataFrame
data = pd.read_csv("Skills Test Data Set - CanvasserData.csv")

# 1. How many unique canvassers did the organization have during their canvassing program?
unique_canvassers = data['Canvasser'].nunique()

# 2. How many shifts did each canvasser work?
shifts_worked = data['Canvasser'].value_counts().reset_index()
shifts_worked.index = shifts_worked.index + 1  # Start the index at 1
shifts_worked.columns = ['Canvasser', 'Shifts Worked']

# 3. How many doors did each canvasser knock and what were the results of those knocks?
doors_knocked_results = data.groupby('Canvasser').sum().reset_index()
doors_knocked_results.index = doors_knocked_results.index + 1  # Start the index at 1
doors_knocked_results.columns = ['Canvasser', 'Not Home', 'Refused', 'Moved', 'Deceased', 'Canvassed']

# 4. What is the contact rate for each canvasser?
# Assuming 'Not Home', 'Refused', 'Moved', and 'Deceased' are considered unsuccessful knocks
data['Unsuccessful Knocks'] = data['Not Home'] + data['Refused'] + data['Moved'] + data['Deceased']
contact_rate_df = data.groupby('Canvasser').apply(lambda x: (x['Canvassed'] / (x['Canvassed'] + x['Unsuccessful Knocks'])).mean()).reset_index()
contact_rate_df.index = contact_rate_df.index + 1  # Start the index at 1
contact_rate_df.columns = ['Canvasser', 'Contact Rate']

# Save the DataFrames as CSV files
pd.DataFrame({'Unique Canvassers': [unique_canvassers]}).to_csv('unique_canvassers.csv', index=False)
shifts_worked.to_csv('shifts_worked.csv', index=False)
doors_knocked_results.to_csv('doors_knocked_results.csv', index=False)
contact_rate_df.to_csv('contact_rate.csv', index=False)

print("CSV files have been created.")
```
